**Задача 3.1.2 Обработка полей**

```python
# Нормализую название полей
train.columns = list_field_train.keys()
```

```python
# Удаление пропусков
train = train.dropna(subset=['diabetes'])
```

```python
# Удаление поля
train = train.drop('unnamed', axis=1)
```

```python
# Сдвигаем показатель на 2 знака чтобы вывести потом возраст
train['age'] = train['age'] * 100
```

```python
# Меняем тип данных у полей
train = train.astype({'age': 'int'})
train = train.astype({'diabetes': 'bool'})
train = train.astype({'smoking': 'bool'})
train = train.astype({'obesity': 'bool'})
train = train.astype({'ac': 'bool'})
train = train.astype({'php': 'bool'})
train = train.astype({'mu': 'bool'})
train = train.astype({'har': 'bool'})
```

**Задача 3.1.4 Проверка категориальных данных**

```python
def list_unique_values_categorical_features(dataset: object) -> str:
    # Выводит список уникальных значений категориальных признаков
    cat_col_names = dataset.select_dtypes(['object', 'bool']).columns.tolist()
    for col in cat_col_names:
        print(f"Категория: [{col}]:")
        print(f"Уникальные значения категории:")
        print(dataset[col].unique())
        print(f"Количество значений по уникальности в категории:")
        print(dataset[col].value_counts())
        print('----------------------------------------------------<NEXT>-')
```
