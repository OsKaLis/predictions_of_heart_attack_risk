**Задача 4.1 Создаю функцию для показы графиков**

```python
class GraphicallyFrequencyIndicators:
    # Класс нужен чтоб строить по признакам частотные графики 

    def __init__(
        self,
        data: object, 
        names_cols: dict, 
    ):
        # Дата сет над которым нужно построить графики по всем признакам
        self.data = data.copy()
        # Словарь наименование колонок
        self.names_cols = names_cols
        # Получаем все количественые столбцы без ID
        self.quantitative_features = self.num_list_names()
        # Получить все категориальные столбцы
        self.categorical_features = self.cal_list_names()
        # Указываем тип графика
        self.type_graph = ''
        # Количество разделений обекта
        self.divided_into = 10
        # Кординаты графика
        self.by_x = 0
        self.by_y = 0
        # Количество графиков
        self.quantity_graphs = 0
        # Показатели для несколько графиков
        self.fig = ''
        self.axes = ''

    def osnova_dly_graphs(self):
        # Подготовка для графиков
        if self.quantity_graphs == 1:
            plt.subplot(1, 2, 1)
        else:
            self.fig, self.axes = plt.subplots(
                nrows=self.quantity_graphs,
                ncols=2,
                figsize=(15, 20),
            )

    def graph_next(self):
        # Следующий график
        self.by_y += 1
        if self.by_y > 1:
            self.by_y = 0
            self.by_x += 1

    def kordinati_default(self):
        # Сброс кординат на старт
        self.by_x = 0
        self.by_y = 0

    def graph_display(
        self,
        columns_list: list,
        distribution:str,
    ):
        # Вывожу графики согласно списку столбцов
        for x in columns_list:
            if distribution == 'categorical_graphs':
                df = self.data[x].value_counts()
            else:
                df = self.data[x]
            if self.type_graph == 'hist':
                if self.quantity_graphs == 1:
                    graph = df.plot(
                        kind='hist',
                        bins=60,
                    )
                else:
                    graph = df.plot(
                        kind='hist',
                        bins=60,
                        ax = self.axes[self.by_x, self.by_y],
                    )
            if self.type_graph in ['bar_value_counts', 'bar']:
                if self.type_graph == 'bar_value_counts':
                    df = df.value_counts()
                if self.quantity_graphs == 1:
                    graph = df.plot(
                        kind='bar',
                        rot=0,
                    )
                else:
                    graph = df.plot(
                        kind='bar',
                        rot=0,
                        ax = self.axes[self.by_x, self.by_y],
                    )
            graph.set_ylabel('Частота')
            graph.set_xlabel(self.names_cols[x])
            graph.grid('True')
            graph.set_title(
                f'График по признаку [{self.names_cols[x]}]'
            )
            self.graph_next()

    def categorical_graphs(self, chart_structure: dict):
        #  Ввод категориальных показателей
        self.strou_graphs(chart_structure, 'categorical_graphs',)

    def quantitative_graphs(self, chart_structure: dict):
        # Ввод количественых показателей
        self.strou_graphs(chart_structure, 'quantitative_graphs',)
        
    def strou_graphs(self, chart_structure: dict, tip_sobiraemih_data: str):
        # Подготовка к строительству графиков
        quantity_graph = 0
        for values in chart_structure.values():
            quantity_graph += len(values)
        self.podschet_quantity_graphs(quantity_graph)
        self.osnova_dly_graphs()
        for key, values in chart_structure.items():
            self.type_graph = key
            self.graph_display(values, tip_sobiraemih_data)
        self.kordinati_default()

    def describe(self):
        # Вывод describe о количественыхз данных
        temp_data = self.data.loc[:, self.quantitative_features]
        temp_data.columns = [self.names_cols[x] for x in self.quantitative_features]
        display(temp_data.describe())

    def output_indicator_chart(
        self,
        target_attribute: str,  # Целевой признак для групировки
        list_num_manes_cols: list,  # Список полей количественных данных для анализа
    ):
        """
        Функция выводит связь целевым показзателем
        и остальными количествеными параметрами
        """
        list_num_manes_cols.append(target_attribute)
        temp_data = self.data.loc[:, list_num_manes_cols]
        temp_data.columns = [self.names_cols[x] for x in list_num_manes_cols]
        sns.pairplot(
            temp_data,
            hue=self.names_cols[target_attribute],
            #corner=True,
        );

    def shkaga_shaped_grouping(self, signs: list, distributors: list):
        """
        Метод перебирает [signs] различные показатели к
        дискретному [distributor] распределённому ранжированию
        distributor - Список показателей что имеет некую последовательность
        signs - Список показателей признаков
        """
        self.podschet_quantity_graphs(len(signs))
        fig, axes = plt.subplots(
            nrows=self.quantity_graphs,
            ncols=2,
            figsize=(15, 25),
        )
        for distributor in distributors:
            for sign in signs:
                if len(self.data[sign].unique()) > 10:
                    self.data[sign] = pd.cut(self.data[sign], 5)
                temp_table = self.data.pivot_table(
                    index=distributor,
                    columns=sign,
                    values='id',
                    aggfunc='count'
                )
                temp_table.plot(
                    kind='barh',
                    grid=True,
                    #xlabel=self.names_cols[sign],
                    ylabel=self.names_cols[distributor],
                    title=f"\nГрафик распределения [{sign}] по [{distributor}]",
                    ax = axes[self.by_x, self.by_y],
                );
                self.graph_next()
        self.kordinati_default()
        
    def podschet_quantity_graphs(self, quantity_graphs: int):
        # Вычисляем количество строеных графиков
        if quantity_graphs > 0:
            if quantity_graphs == 1:
                self.quantity_graphs = 1
            if quantity_graphs%2 == 0:
                if quantity_graphs == 2:
                    self.quantity_graphs = 2
                else:
                    self.quantity_graphs = int(quantity_graphs / 2)
            else:
                self.quantity_graphs = int((quantity_graphs + 1) / 2)

    def cal_list_names(self):
        # Получаем список категориальных столбцов
        return self.data.select_dtypes('object').columns.tolist()
    
    def num_list_names(self):
        # Получаем список количественных столбцов
        quantitative_features = self.data.select_dtypes(
            ['float', 'int']
        ).columns.tolist()
        if 'id' in quantitative_features:
            quantitative_features.remove('id')
        return quantitative_features
```

```python
def heat_map_output(
    data: object,         # Дата фрейм
    list_names_col:dict,  # Словарь с претанзиционым наименованием 
):
    """
    Функция строит теплавую карту на основании дата фрейма    
    """
    temp_data = data.drop('id', axis=1)
    
    corrective_list = list()
    for x in temp_data.columns:
        corrective_list.append(list_names_col[x])
    temp_data.columns = corrective_list

    quantitative_signs = (
        temp_data.select_dtypes(['float', 'int']).columns.tolist()
    )
    plt.figure(figsize = (16,16))
    plt.title(
        "Теплавая карта коэффициенты корреляции между всеми признаками в дата фрейме\n",
        fontsize =10
    )
    sns.heatmap(
        temp_data.phik_matrix(interval_cols=quantitative_signs),
        annot=True,
        fmt='.2f'
    );
```

**Задача 6.3. Добавляем ОНЕ и Ordinal кодирование в пайплан**

```python
# Создаём структуру кодирование OHE
ohe_pipe = Pipeline(
    [
        (
            'simpleImputer_ohe',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ohe',
            OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)
        )
    ]
)
```

```python
# Создаём структуру кодирование Ordinal
ord_pipe = Pipeline(
    [
        (
            'simpleImputer_before_ord',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ord',
            OrdinalEncoder(
                categories=[
                    ['Да', 'Нет', 'неизвестно'],
                    ['Да', 'Нет', 'неизвестно'], 
                    ['Да', 'Нет', 'неизвестно'],
                    ['Да', 'Нет', 'неизвестно'],
                    ['Да', 'Нет', 'неизвестно'],
                    ['Да', 'Нет', 'неизвестно'],
                    ['Да', 'Нет', 'неизвестно'],
                ],
                handle_unknown='use_encoded_value',
                unknown_value=np.nan)
        ),
        (
            'simpleImputer_after_ord',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
)
```

**Задача 6.4. Обединяем наши подготовки**

```python
# Обеденим шаги по подготовке данных
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', StandardScaler(), num_columns),
    ],
    remainder='passthrough'
)
```

**Задача 6.5. Cоздание параметров для пайплана**

```python
# Создаём словарь из 4 моделеё с параметрами
param_grid = [
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 11),
        'models__max_features': range(2, 11),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough'],
    },
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2, 11),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough'],
    },
    {
        'models': [SVC(random_state=RANDOM_STATE)],
        'models__C': [0.1, 1, 10, 100],
        'models__gamma': [0.1, 1, 10, 100],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough'],
    },
]
```

**Задача 6.6. Создание итогового пайплана для обучение**

```python
# Создаём итоговый пайплан
pipe_final = Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', DecisionTreeClassifier(random_state=RANDOM_STATE)),
    ]
)
```
